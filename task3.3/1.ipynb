```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

train = pd.read_csv("./fish_train.csv")
test = pd.read_csv("./fish_reserved.csv")

X = train.drop(columns=["Weight"])
y = train["Weight"]

X_num = X.drop(columns=["Species"])
X_cat = pd.get_dummies(X["Species"], drop_first=False)

pca = PCA(n_components=3).fit(X_num)
X_pca = pd.DataFrame(pca.transform(X_num), columns=["PC1","PC2","PC3"])

poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = pd.DataFrame(poly.fit_transform(X_pca), columns=poly.get_feature_names_out(["PC1","PC2","PC3"]))

X_all = pd.concat([X_poly, X_cat], axis=1)

model = LinearRegression().fit(X_all, y)

Xt_num = test.drop(columns=["Species"])
Xt_cat = pd.get_dummies(test["Species"], drop_first=False)

Xt_pca = pd.DataFrame(pca.transform(Xt_num), columns=["PC1","PC2","PC3"])
Xt_poly = pd.DataFrame(poly.transform(Xt_pca), columns=poly.get_feature_names_out(["PC1","PC2","PC3"]))

Xt_all = pd.concat([Xt_poly, Xt_cat], axis=1)
Xt_all = Xt_all.reindex(columns=X_all.columns, fill_value=0)

pred = model.predict(Xt_all)
[float(round(i, 1)) for i in pred]
```
