```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture
import numpy as np
np.random.seed(11)
digits = load_digits()
X = digits.data

pca = PCA(n_components=0.99, svd_solver="full")
X_pca = pca.fit_transform(X)
out1 = X_pca.shape[1]

aics = {}
for n in range(50, 251, 10):
    gmm = GaussianMixture(n_components=n, covariance_type="full", random_state=11)
    gmm.fit(X_pca)
    aics[n] = gmm.aic(X_pca)

best_n = min(aics, key=aics.get)

gmm = GaussianMixture(n_components=best_n, covariance_type="full", random_state=11)
gmm.fit(X_pca)


samples, _ = gmm.sample(100)
X_gen = pca.inverse_transform(samples)

out = round(samples[:, 0].mean(), 3)

(out1, best_n, gmm.converged_, out)
```
