# Task 1

```python
import pandas as pd
```

```python
df = pd.read_csv('/pulsar_stars_exam.csv')
```

```python
df.info()
```

```python
df = df[((df['target'] == 1) & ((df['mip'] >= 8.2421875) & (df['mip'] <= 116.8359375))) | ((df['target'] == 0) & ((df['mip'] >= 62.3984375) & (df['mip'] <= 182.8359375)))]
```

```python
df.count()
```

```python
round(df['ekip'].mean(), 3)
```

```python
df['sip']
```

```python
df = df.sort_values(by='sip', ascending=True, ignore_index=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns='target')
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=9, stratify=df['target'], train_size=0.8, test_size=0.2)
```

```python
round(X_train['ekc'].mean(), 3)
```

```python
round(X_test['ekc'].mean(), 3)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_sc = scaler.transform(X_train)
X_test_sc = scaler.transform(X_test)

X_train_sc = pd.DataFrame(X_train_sc, columns=X_train.columns)
X_test_sc = pd.DataFrame(X_test_sc, columns=X_test.columns)
X_train_sc
```

```python
round(X_train_sc['mc'].mean(), 3)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=9)
lr.fit(X_train_sc, y_train)
y_pred1 = lr.predict(X_test_sc)
```

```python
from sklearn.metrics import confusion_matrix, f1_score
tn, fp, fn, tp = confusion_matrix(y_test, y_pred1).ravel().tolist()
tn, fp, fn, tp
```

```python
round(f1_score(y_test, y_pred1), 3)
```

```python
from sklearn.neighbors import KNeighborsClassifier

ml = KNeighborsClassifier(n_neighbors=3)
ml.fit(X_train_sc, y_train)
y_pred2 = ml.predict(X_test_sc)
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, y_pred2).ravel().tolist()
tn, fp, fn, tp
```

```python
round(f1_score(y_test, y_pred2), 3)
```

# Task 2

```python
test = pd.read_csv('/content/MNIST_test.csv')
train = pd.read_csv('/content/MNIST_train.csv')
```

```python
train = train[(train['label'] == 0) | (train['label'] == 9)]
test = test[(test['label'] == 0) | (test['label'] == 9)]
```

```python
X_train = train.drop(columns='label')
y_train = train['label']
X_test = test.drop(columns='label')
y_test = test['label']
```

```python
from sklearn.decomposition import PCA
pca = PCA(svd_solver='full')
pca.n_components = 25
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
sum(pca.explained_variance_ratio_)
```

```python
round(X_train_pca[7, 1], 3)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
bg = BaggingClassifier(DecisionTreeClassifier(class_weight='balanced'),max_samples=0.5, max_features=0.5,bootstrap=False,random_state=31)

bg.fit(X_train_pca, y_train)
y_pred = bg.predict(X_test_pca)
round(bg.score(X_test_pca, y_test), 3)
```

```python
tn, fp, fn, tp = confusion_matrix(y_pred, y_test).ravel().tolist()
(tn, fp, fn, tp)
```
