```python
import numpy as np
from sklearn.cluster import MiniBatchKMeans
from PIL import Image

img = Image.open("image.jpg").convert("RGB")
img_np = np.array(img)

mean_before = round(img_np.mean(), 3)

img_norm = img_np / 255.0
mean_norm = round(img_norm.mean(), 3)

h, w, c = img_norm.shape
pixels = img_norm.reshape(-1, 3)

kmeans = MiniBatchKMeans(n_clusters=16, random_state=10)
labels = kmeans.fit_predict(pixels)

new_pixels = kmeans.cluster_centers_[labels]
new_img = new_pixels.reshape(h, w, 3)

mean_after = round(new_img.mean(), 3)

out_img = (new_img * 255).astype(np.uint8)
Image.fromarray(out_img).save("compressed_16_colors.png")

(mean_before, mean_norm, mean_after)
```
