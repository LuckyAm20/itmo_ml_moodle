```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

data = pd.DataFrame([
    [1, 36, 11, 1],
    [2, 76, 14, 0],
    [3, 30, 40, 0],
    [4, 12, 17, 1],
    [5, 97, 28, 1],
    [6, 98, 57, 0],
    [7, 40, 24, 1],
    [8, 53, 69, 1],
    [9, 55, 45, 0],
    [10, 60, 43, 1]
], columns=["id", "X", "Y", "Class"])

x0, y0 = 63, 72

points = data[["X","Y"]].values
target = np.array([x0, y0])

euclid = np.sqrt(((points - target)**2).sum(axis=1))
manhattan = np.abs(points - target).sum(axis=1)

nearest_euclid_dist = round(euclid.min(), 3)
nearest_manhattan_dist = round(manhattan.min(), 3)

idx_sorted_euclid = data.iloc[np.argsort(euclid)]["id"].values[:3]
idx_sorted_manhattan = data.iloc[np.argsort(manhattan)]["id"].values[:3]

knn_e = KNeighborsClassifier(n_neighbors=3, p=2)
knn_e.fit(data[["X","Y"]], data["Class"])
class_e = int(knn_e.predict([[x0, y0]])[0])

knn_m = KNeighborsClassifier(n_neighbors=3, p=1)
knn_m.fit(data[["X","Y"]], data["Class"])
class_m = int(knn_m.predict([[x0, y0]])[0])

nearest_euclid_dist, idx_sorted_euclid, class_e, nearest_manhattan_dist, idx_sorted_manhattan, class_m
```
