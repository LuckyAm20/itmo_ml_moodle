```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.DataFrame({
    "X":[8,10,4,24,13,16,5,3,1,18],
    "Y":[16,21,12,47,35,32,11,8,4,38]
})

X = df[['X']]
y = df['Y']

model = LinearRegression().fit(X, y)

meanX = round(df['X'].mean(), 2)
meanY = round(df['Y'].mean(), 2)
theta1 = round(model.coef_[0], 3)
theta0 = round(model.intercept_, 3)
r2 = round(model.score(X, y), 3)

meanX, meanY, theta1, theta0, r2
```
