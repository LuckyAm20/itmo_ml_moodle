```python
from sklearn.model_selection import train_test_split
```

```python
sip = df_new['sip'].sort_values(ascending=True)
sip
```

```python
df_new_sort = df_new.copy()
df_new_sort['sip']
```

```python
df_new_sort.sort_values(by='sip', ascending=True, ignore_index=True, inplace=True)
df_new_sort['sip']
```

```python
train, test = train_test_split(df_new_sort, test_size=0.2, train_size=0.8, random_state=33, stratify=df_new_sort['target'])
```

```python
train
```

```python
round(train['stdc'].max(), 3)
```

```python
X_train = train.drop(columns=['target'])
y_train = train['target']

X_test = test.drop(columns=['target'])
y_test = test['target']
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
X_test_scaled = pd.DataFrame(
    X_test_scaled,
    columns=X_test.columns,
    index=X_test.index
)
X_train_scaled = pd.DataFrame(
    X_train_scaled,
    columns=X_train.columns,
    index=X_train.index
)
```

```python
round(X_train_scaled['stdip'].mean(), 3)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train_scaled, y_train)
y_new = lr.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_new)
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, y_new).ravel().tolist()
tp
```

```python
from sklearn.metrics import f1_score
round(f1_score(y_test, y_new), 3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train_scaled, y_train)
y_new = neigh.predict(X_test_scaled)
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, y_new).ravel().tolist()
tp
```

```python
from sklearn.metrics import f1_score
round(f1_score(y_test, y_new), 3)
```

# Task 2

```python
import pandas as pd

train = pd.read_csv('/MNIST_train.csv')
test = pd.read_csv('/MNIST_test.csv')
train.shape
```

```python
train = train[(train['label'] == 2) | (train['label'] == 9)]
test = test[(test['label'] == 2) | (test['label'] == 9)]
```

```python
test.shape
```

```python
train['label'].value_counts()
```

```python
X_train = train.drop(columns=['label'])
y_train = train['label']

X_test = test.drop(columns=['label'])
y_test = test['label']
```

```python
from sklearn.decomposition import PCA

pca = PCA(svd_solver='full', n_components=40)
pca.fit(X_train)
sum(pca.explained_variance_ratio_)
```

```python
X_pca2 = pca.fit_transform(X_train)
pd.DataFrame(X_pca2).head(n=10)
```

```python
X_pca2_test = pca.transform(X_test)
pd.DataFrame(X_pca2_test).head()
```

```python
round(X_pca2[9, 8], 3)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

ada = AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=3, class_weight='balanced'),n_estimators=300,learning_rate=0.5,random_state=3)
ada.fit(X_pca2, y_train)
y_pred = ada.predict(X_pca2_test)
```

```python
round(ada.score(X_pca2_test, y_test), 3)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel().tolist()
tp
```
